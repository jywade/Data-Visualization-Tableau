### Date Part (Splitting Dates)

```python
nf[["Date Part 1", "Date Part 2"]]  = nf["date_added"].str.split(", ", expand=True)
nf
```

### renaming columns

```python
nf.rename(columns={"Date Part 2": "Year"}, inplace=True)
nf.rename(columns={"Date Part 1": "Month&Day"}, inplace=True)
nf
```

### Replacing Null Values

```python
nf["cast"].fillna(value="no cast", inplace=True)
nf
```

### Lambda Functions

```python
the_adder = lambda x: x + 2393
the_adder(5728)
```

```python
nf["genre_count"] = nf["listed_in"].map(lambda x: len(x.split(",")))      
nf
    
    # creates new column showing the count of genres in the 'listed_in' columnn
```

### Group By and Sort

```python
nf_movies_by_country = nf.groupby("country")["show_id"].count().reset_index()
nf_movies_by_country.sort_values(by="show_id", ascending=False)
```

### Joining Tables 

```python
nf_merge = pd.merge(left=nf2, right=nf_movies_by_country, how="inner", left_on=["country"], right_on=["country"])
nf_merge
```
